In machine translation scenarios both encoder and decoder are LSTM.

A seq2seq consists of the Encoder-Decoder structure. It first work with the encoder encoding all of the input, turning the inputs
into a single representation called as Context Vector. 
It contains all the captured information whatever the encoder was able to capture. 

This context vector is then sent to the Decoder to produce the output. In machine translation scenario bot the encoder and decoder are recurrent neural networks.(RNN)

(LSTM cells in practice.)



In real world scenarios, length of the context vectors are usually 256 & 512.



The encoder is confined to send a single vector. Limitation of Seq2Seq. Problems of long sequences. Large number of hidden units.
Model overfits with short sequence.

Problem that attention solves

Seq2Seq with Attention model

Attention model with all the hidden states processed so far. Advantage of having variable length context vector.



Attention Decoder.
it pays attention to the appropriate part of input sequence. How?

Summary> A Seq2Seq model works by feeding 1 element of the input sequence at a time to the encoder.

1. Fixed size of context matrix passed from the encoder to the decoder which is a bottleneck
